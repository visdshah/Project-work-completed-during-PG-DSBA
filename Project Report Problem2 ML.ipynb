```python
import string
nltk.download('stopwords')       #downloaded for stopwords
nltk.download('punkt')           #downloaded for punctuations
```

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)
print(stopwords)
```

```python
all_Words= pd.DataFrame(all_Words)
all_Words.columns = ['Content']
```

```python
stop = stopwords
all_Words['Cleaned'] = all_Words['Content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
all_Words['Cleaned'].head(30)
```

```python
# converting all words into lowercase
all_Words['Cleaned']= all_Words['Cleaned'].apply(lambda x: 'np.nan'.join(x.lower() for x in x.split()))
all_Words['Cleaned']
```

```python
import numpy as np 
```

```python
all_Words['Cleaned']= all_Words['Cleaned'].replace('--', float('NaN'), regex = True)
```

```python
all_Words['Cleaned']= all_Words['Cleaned'].replace(r'^s*$', float('NaN'), regex = True)
all_Words = all_Words['Cleaned'].dropna()
all_Words.head(40)
```

```python
# Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
all_Words= all_Words.apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
all_Words.head(50)
```

```python
freq = pd.Series(' '.join(all_Words).split()).value_counts()[:40]
freq
```

```python
roose = inaugural.raw('1941-Roosevelt.txt')
roose =pd.read_csv('C:\\Users\\Vishal.Shah\\Desktop\\Roose.csv', sep='delimiter', header=None,engine='python')
roose=[x for x in pd.Series(' '.join(roose[0]).split())]
roose = pd.DataFrame(roose)
roose.columns = ['content']
roose['content']= roose['content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
roose['content'] = roose['content'].str.replace('[^\w\s]','')
roose['content']= roose['content'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
roose['content']= roose['content'].replace(r'^s*$', float('NaN'), regex = True)
roose['content']= roose['content'].replace('--', float('NaN'), regex = True)
roose = roose['content'].dropna()
roose_freq = pd.Series(' '.join(roose).split()).value_counts()[:40]
roose_freq
```

```python
# Word Cloud 
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='black', 
                min_font_size = 10, random_state=100).generate(wc_a) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for Roosevelt's speech (after cleaning)!!")
plt.show()

```
